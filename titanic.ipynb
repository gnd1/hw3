```python
import pandas as pd

location = "titanic3.xls"
df = pd.read_excel(location)
df.head()
```

```python
total_passengers = len(df) 

total_women = len(df[df['sex'] == 'female'])
total_men = len(df[df['sex'] == 'male'])
total_survived = len(df['survived'])
total_over18 = len(df[df['age'] >= 18])
total_under18 = len(df[df['age'] < 18])

print(total_women) 
print(total_men)
print(total_survived)
print(total_over18)
print(total_under18)
```

```python
womensurvived = len(df[(df['sex'] == 'female') & (df['survived'] == 1)]) 
mensurvived = len(df[(df['sex'] == 'male') & (df['survived'] == 1)]) 
fcsurvived = len(df[(df['pclass'] == 1) & (df['survived'] == 1)]) 
scsurvived = len(df[(df['pclass'] == 2) & (df['survived'] == 1)]) 
tcsurvived = len(df[(df['pclass'] == 3) & (df['survived'] == 1)]) 
adultsurvived = len(df[(df['age'] >= 18) & (df['survived'] == 1)]) 
childsurvived = len(df[(df['age'] < 18) & (df['survived'] == 1)]) 
womendied = len(df[(df['sex'] == 'female') & (df['survived'] == 0)]) 
fcwomendied = len(df[(df['sex'] == 'female') & (df['survived'] == 0) & (df['pclass'] == 1)])
mendied = len(df[(df['sex'] == 'male') & (df['survived'] == 0)]) 
tcmendied = len(df[(df['sex'] == 'male') & (df['survived'] == 0) & (df['pclass'] == 3)])
childdied = len(df[(df['age'] < 18) & (df['survived'] == 0)])
malchilddied = len(df[(df['sex'] == 'male') & (df['survived'] == 0) & (df['age'] < 18)])
tcmalchilddied = len(df[(df['sex'] == 'male') & (df['survived'] == 0) & (df['age'] < 18) & (df['pclass'] == 3)])

print(womensurvived)
print(mensurvived)
print(fcsurvived)
print(scsurvived)
print(tcsurvived)
print(adultsurvived)
print(childsurvived)
print(womendied)
print(fcwomendied)
print(mendied)
print(tcmendied)
print(childdied)
print(malchilddied)
print(tcmalchilddied)
```

```python
p_women_survived = womensurvived / total_women 
p_men_survived = mensurvived / total_men 
p_fc_survived = fcsurvived / total_survived 
p_sc_survived = scsurvived / total_survived 
p_tc_survived = tcsurvived / total_survived 
p_adult_survived = adultsurvived / total_over18 
p_child_survived = childsurvived / total_under18 
p_fc_women_died = fcwomendied / womendied 
p_tc_men_died = tcmendied / mendied
p_mal_child_died = malchilddied / childdied 
p_tcmal_child_died = tcmalchilddied / malchilddied 

print(p_women_survived)
print(p_men_survived)
print(p_sc_survived)
print(p_tc_survived)
print(p_adult_survived)
print(p_child_survived)
print(p_fc_women_died)
print(p_tc_men_died)
print(p_mal_child_died)
print(p_tcmal_child_died)
```
